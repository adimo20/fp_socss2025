# Untersuchung der Extraktionsqualität

Im Folgenden wird demonstriert wie wir sicherstellen können, wie gut unsere Textextraktion aus den Volltexten funtkioniert hat. Um dies zu tun laden wir zunächst ein Sample von Daten die wir in einem vorherigen Schritt extrahiert haben. Dabei enthält der Datensatz `data` den wir verwenden den extrahierten Text Teil in der Spalte `text` und den originalen Volltext `plainpagefulltext`. 

```python
import pandas as pd 
from scripts.information_extractor_lib import ExtractionValidator

data = pd.read_pickle("dat_joined.pkl")
data.head(10)
```

Die `find_similiar_sequence` iluustriert wie die `ExtractionValidator` die Accuracy der Textextraktion berechnet. Es wird der extrahierte Text mit dem original Text vergleichen. Der extrahierte Text läuft wie ein Fenster über den originalen Text. Wenn der extrahierte Text und zu vergleichende String aus dem orignal Dokument eine (levenstein) string similiarity von > threshold (in diesem Fall 0.95) haben wir der extrahierte Text als korrekt extrahiert angesehen. Wobei bei der Berechnung der Accuracy nicht dieser vollständige Prozess zur Anwendung kommt. es wird zunächst überprüft ob eine 1:1 Übereinstimmung vorhanden ist. Wenn das nicht der Fall ist, wird die Methode `find_similiar_sequence` angewendet, wobei diese nur überprüft ob mindestens ein Wert den Threshold überschreitet und dann abbricht (aus Gründen der Recheneffizenz).

```python
extracted_text = data.text.tolist()
full_text = data.plainpagefulltext.tolist()

print(extracted_text[0])
print(full_text[0])
```

```python
validator = ExtractionValidator()
result_df = validator.find_similiar_sequence(dst_doc=extracted_text[0], source_doc=full_text[0], threshold=0.95, break_when_threshold=False)
result_df
```

```python
validator2 = ExtractionValidator()
is_match_bool = validator2.is_match(dst_doc=extracted_text[0], source_doc=full_text[0], threshold=0.95)
is_match_bool
```

```python

validator3 = ExtractionValidator(dst_doc=extracted_text, source_doc=full_text, threshold=0.95)

extraction_accuracy = validator3.calculate_extraction_accuracy()

print(f"Extraction accurary is: {round(extraction_accuracy[1], 3)}")

```
